# heat map 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Define the name for the new folder
output_folder = 'output_visuals'

# Define the name of your data file
csv_file = "Concrete_Data_Yeh.csv"

# --- Main script ---
try:
    # Create the directory; `exist_ok=True` prevents an error if it already exists
    os.makedirs(output_folder, exist_ok=True)
    
    # Let pandas automatically read the header from the first row of the CSV
    data = pd.read_csv(csv_file)
    
    # Calculate the correlation matrix
    correlation_matrix = data.corr()
    
    print("--- Correlation Matrix ---")
    print(correlation_matrix)
    
    # Set up the plot
    plt.figure(figsize=(12, 10))
    
    # Generate the heatmap
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
    
    # Add a title and adjust layout
    plt.title('Correlation Heatmap of Concrete Components')
    plt.tight_layout()
    
    # Define the full path for the output file
    file_path = os.path.join(output_folder, 'correlation_heatmap.svg')
    
    # Save the figure as an SVG file
    plt.savefig(file_path, format='svg')
    
    print(f"\n✅ Heatmap successfully saved to: '{file_path}'")

except FileNotFoundError:
    print(f"\n❌ ERROR: The file '{csv_file}' was not found.")
    print("Please make sure the file is in the same directory as the script.")
except Exception as e:
    print(f"\nAn unexpected error occurred: {e}")
```
